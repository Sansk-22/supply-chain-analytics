```python
import pandas as pd

df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin-1')

print(df.shape)
print(df.columns.tolist())
print(df.head())
```

```python
print(df.columns.tolist())
print("\nNull values:")
print(df.isnull().sum())
print("\nData types:")
print(df.dtypes)
```

```python



df = df.drop(columns=['Product Description', 'Product Image', 
                        'Customer Password', 'Customer Email',
                        'Customer Street'])


df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])


df['Customer Zipcode'] = df['Customer Zipcode'].fillna(0)


print("Shape after cleaning:", df.shape)
print("\nRemaining null values:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("\nDate column type now:", df['order date (DateOrders)'].dtype)
```

```python

print("=== DELIVERY STATUS ===")
print(df['Delivery Status'].value_counts())


print("\n=== MARKETS ===")
print(df['Market'].value_counts())


print("\n=== CUSTOMER SEGMENTS ===")
print(df['Customer Segment'].value_counts())


print("\n=== SHIPPING MODES ===")
print(df['Shipping Mode'].value_counts())

print("\n=== TOP CATEGORIES ===")
print(df['Category Name'].value_counts().head(10))
```

```python


print("=== LATE DELIVERY BY MARKET ===")
late_by_market = df[df['Delivery Status'] == 'Late delivery'].groupby('Market').size()
total_by_market = df.groupby('Market').size()
late_pct_market = (late_by_market / total_by_market * 100).round(2)
print(late_pct_market.sort_values(ascending=False))


print("\n=== LATE DELIVERY BY SHIPPING MODE ===")
late_by_ship = df[df['Delivery Status'] == 'Late delivery'].groupby('Shipping Mode').size()
total_by_ship = df.groupby('Shipping Mode').size()
late_pct_ship = (late_by_ship / total_by_ship * 100).round(2)
print(late_pct_ship.sort_values(ascending=False))


print("\n=== AVG SHIPPING DELAY BY MARKET ===")
df['Shipping Delay'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
print(df.groupby('Market')['Shipping Delay'].mean().round(2).sort_values(ascending=False))


print("\n=== PROFIT BY CATEGORY ===")
print(df.groupby('Category Name')['Order Profit Per Order'].mean().round(2).sort_values(ascending=False).head(10))


print("\n=== PROFIT BY MARKET ===")
print(df.groupby('Market')['Order Profit Per Order'].mean().round(2).sort_values(ascending=False))
```

```python
# Recheck Late Delivery by Shipping Mode properly
print("=== TOTAL ORDERS BY SHIPPING MODE ===")
print(df.groupby('Shipping Mode').size())

print("\n=== LATE ORDERS BY SHIPPING MODE ===")
print(df[df['Delivery Status'] == 'Late delivery'].groupby('Shipping Mode').size())

print("\n=== DETAILED BREAKDOWN ===")
cross = pd.crosstab(df['Shipping Mode'], df['Delivery Status'])
print(cross)

print("\n=== PERCENTAGE BREAKDOWN ===")
cross_pct = pd.crosstab(df['Shipping Mode'], df['Delivery Status'], normalize='index') * 100
print(cross_pct.round(2))
```

```python
# Step 5: Sales & Profit Analysis

# 1. Monthly sales trend
print("=== MONTHLY SALES TREND ===")
df['Month'] = df['order date (DateOrders)'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum().round(2)
print(monthly_sales.tail(12))

# 2. Discount impact on profit
print("\n=== DISCOUNT IMPACT ON PROFIT ===")
df['Discount Category'] = pd.cut(df['Order Item Discount Rate'], 
                                   bins=[0, 0.05, 0.10, 0.20, 1.0],
                                   labels=['0-5%', '5-10%', '10-20%', '20%+'])
discount_profit = df.groupby('Discount Category', observed=True)['Order Profit Per Order'].mean().round(2)
print(discount_profit)

# 3. Top 10 most profitable products
print("\n=== TOP 10 PROFITABLE PRODUCTS ===")
print(df.groupby('Product Name')['Order Profit Per Order'].mean().round(2).sort_values(ascending=False).head(10))

# 4. Customer segment profit
print("\n=== PROFIT BY CUSTOMER SEGMENT ===")
print(df.groupby('Customer Segment')['Order Profit Per Order'].mean().round(2).sort_values(ascending=False))

# 5. Order status breakdown
print("\n=== ORDER STATUS ===")
print(df['Order Status'].value_counts())
```

```python

df.to_csv('supply_chain_cleaned.csv', index=False)


late_ship = pd.crosstab(df['Shipping Mode'], 
                         df['Delivery Status'], 
                         normalize='index') * 100
late_ship.to_csv('late_delivery_shipping.csv')


monthly_sales.to_csv('monthly_sales.csv')


profit_category = df.groupby('Category Name')['Order Profit Per Order'].mean().round(2)
profit_category.to_csv('profit_by_category.csv')


profit_market = df.groupby('Market')['Order Profit Per Order'].mean().round(2)
profit_market.to_csv('profit_by_market.csv')

print("All files saved successfully!")
print("\nFiles created:")
print("1. supply_chain_cleaned.csv â main dataset for Power BI")
print("2. late_delivery_shipping.csv â shipping analysis")
print("3. monthly_sales.csv â sales trend")
print("4. profit_by_category.csv â category profits")
print("5. profit_by_market.csv â market profits")

order_status = df['Order Status'].value_counts().reset_index()
order_status.columns = ['Order Status', 'Count']
order_status.to_csv('order_status.csv', index=False)
print(" Order status file saved!")

discount_profit = df.groupby('Discount Category', observed=True)['Order Profit Per Order'].mean().round(2).reset_index()
discount_profit.columns = ['Discount Category', 'Avg Profit']
discount_profit.to_csv('discount_impact.csv', index=False)
print("Discount impact file saved!")
```
